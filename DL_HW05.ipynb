# iris資料集介紹：

```python
from keras.utils import np_utils
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

資料集內容：150朵花的特徵值(sepal length, sepal width, petal length, petal width)

類別：150朵花的屬性('setosa', 'versicolor', 'virginica')


# 資料預處理(data preprocessing)

```python
a=np.random.permutation(150)
b=np.random.permutation(125)
x_train_data=iris.data[a]
y_train_label=iris.target[a]
x_test_data=iris.data[b]
y_test_label=iris.target[b]
```

```python
print(x_test_data)
```

```python
print('x_train:',len(x_train_data),
      'x_test:',len(x_test_data),
      'y_train:',len(y_train_label),
      'y_test:',len(y_test_label)
     )
```

```python
x_train =x_train_data.reshape(150, 4).astype('float32')
x_test = x_test_data.reshape(125, 4).astype('float32')
```

```python
print(x_test)
```

```python
x_train_normalize = x_train_data / 8
x_test_normalize = x_test_data / 8
```

```python
print(x_test_normalize)
```

```python
y_train_OneHot = np_utils.to_categorical(y_train_label)
y_test_OneHot = np_utils.to_categorical(y_test_label)
```

```python
print(y_test_OneHot)
```

# 建立模型

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(units=600, 
                input_dim=4, 
                kernel_initializer='normal', 
                activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(units=3, 
                kernel_initializer='normal', 
                activation='softmax'))
```

```python
print(model.summary())
```

# 訓練模型

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

```python
train_history=model.fit(x=x_train_normalize,
                        y=y_train_OneHot,validation_split=0.2, 
                        epochs=10, batch_size=10,verbose=2)
```
